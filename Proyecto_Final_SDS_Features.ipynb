### Creación de nuevas variables

```python
# Convertir el valor de la fecha hora en variables separadas
df = df.copy()
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['hour'] = df['trans_date_trans_time'].dt.hour
df['minute'] = df['trans_date_trans_time'].dt.minute
df['second'] = df['trans_date_trans_time'].dt.second
df['day'] = df['trans_date_trans_time'].dt.day
df['month'] = df['trans_date_trans_time'].dt.month
df['year'] = df['trans_date_trans_time'].dt.year
```

```python
# Ordenar el dataset
df_sorted = df.sort_values(by=['cc_num', 'trans_date_trans_time']).copy()

df_sorted['distance_to_previous'] = 0.0

# Calcular la distancia a la transaccion previa
for i in range(1, len(df_sorted)):
    if df_sorted['cc_num'].iloc[i] == df_sorted['cc_num'].iloc[i-1]:
        coord_prev = (df_sorted['merch_lat'].iloc[i-1], df_sorted['merch_long'].iloc[i-1])
        coord_curr = (df_sorted['merch_lat'].iloc[i], df_sorted['merch_long'].iloc[i])
        distance = geodesic(coord_prev, coord_curr).kilometers
        df_sorted.at[i, 'distance_to_previous'] = distance
```

```python
df = df_sorted
```

```python
# Ordenar el dataset
df_sorted = df.sort_values(by=['cc_num', 'trans_date_trans_time']).copy()

df_sorted['distance_to_owner'] = 0.0

for i in range(1, len(df_sorted)):
    if df_sorted['cc_num'].iloc[i] == df_sorted['cc_num'].iloc[i-1]:
        coord_prev = (df_sorted['lat'].iloc[i-1], df_sorted['long'].iloc[i-1])
        coord_curr = (df_sorted['merch_lat'].iloc[i], df_sorted['merch_long'].iloc[i])
        distance = geodesic(coord_prev, coord_curr).kilometers
        df_sorted.at[i, 'distance_to_owner'] = distance
```

```python
df = df_sorted
```

```python
# Ordenar el dataset
df_sorted = df.sort_values(by=['cc_num', 'trans_date_trans_time']).copy()

categories = df_sorted['category'].unique()

# Por cada una de las categorias calcular la suma acumulatoria de cada una de las tarjetas
for category in categories:
    df_sorted[category] = 0

for cc_num, group in df_sorted.groupby('cc_num'):
    counts = {category: 0 for category in categories}
    for index, row in group.iterrows():
        current_category = row['category']
        for category in categories:
            df_sorted.at[index, category] = counts[category]
        counts[current_category] += 1
```

```python
df['transaction_date'] = pd.to_datetime(df['unix_time'], unit='s')

# Características temporales adicionales
df['part_of_day'] = pd.cut(df['transaction_date'].dt.hour,
                           bins=[0, 6, 12, 18, 24],
                           labels=['night', 'morning', 'afternoon', 'evening'],
                           right=False).astype(str)
df['month_of_year'] = df['transaction_date'].dt.month
df['day_of_month'] = df['transaction_date'].dt.day

# Tiempo desde la última transacción
df['prev_unix_time'] = df.groupby('cc_num')['unix_time'].shift(1)
df['time_since_last_transaction'] = df['unix_time'] - df['prev_unix_time']

# Promedio de gasto mensual y diario
df['average_monthly_spend'] = df.groupby(['cc_num', 'year', 'month'])['amt'].transform('mean')
df['average_daily_spend'] = df.groupby(['cc_num', 'year', 'month', 'day'])['amt'].transform('mean')

# Varianza de gasto mensual
df['monthly_spend_variance'] = df.groupby(['cc_num', 'year', 'month'])['amt'].transform('var')

# Frecuencia de transacciones en comercios únicos
df['unique_merchants_per_month'] = df.groupby(['cc_num', 'year', 'month'])['merch_lat'].transform('nunique')

# Frecuencia de transacciones en una ubicación
df['transactions_per_location'] = df.groupby(['cc_num', 'lat', 'long'])['unix_time'].transform('count')

# Velocidad promedio de desplazamiento
df['prev_lat'] = df.groupby('cc_num')['lat'].shift(1)
df['prev_long'] = df.groupby('cc_num')['long'].shift(1)
df['prev_time'] = df.groupby('cc_num')['unix_time'].shift(1)
df['distance_traveled'] = np.sqrt((df['lat'] - df['prev_lat'])**2 + (df['long'] - df['prev_long'])**2)
df['time_traveled'] = df['unix_time'] - df['prev_time']
df['average_speed'] = df['distance_traveled'] / df['time_traveled']

# Verificación de columnas de categorías y cálculo de categoría predominante
category_columns = [col for col in df.columns if col.startswith('category_')]
if category_columns:
    df['predominant_category'] = df[category_columns].idxmax(axis=1).astype(str)
    df['prev_category'] = df.groupby('cc_num')['predominant_category'].shift(1).astype(str)
    df['category_change'] = (df['predominant_category'] != df['prev_category']).astype(int)
else:
    df['predominant_category'] = 'none'
    df['prev_category'] = 'none'
    df['category_change'] = 0

# Codificación one-hot de variables categóricas
df = pd.get_dummies(df, columns=['part_of_day', 'predominant_category'], drop_first=True)

# Eliminamos columnas no necesarias
columns_to_drop = ['cc_num', 'unix_time', 'zip', 'lat', 'long', 'merch_lat', 'merch_long', 'amt_month', 'amt_year', 'transaction_date', 'prev_unix_time', 'prev_category']
df.drop(columns=columns_to_drop, inplace=True)

# Rellenar valores NaN que pueden haber sido introducidos por el método shift()
df.fillna(0, inplace=True)
```
